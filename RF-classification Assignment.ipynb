```python
import pandas as pd
```

```python
dataset=pd.read_csv("kidney_disease.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
dataset['classification_ckd\t'].value_counts()
```

```python
independent=dataset[["id","age","bp","sg","al","su","bgr","bu","sc","sod",'dm_ yes', 'dm_no', 'dm_yes', 'cad_no', 'cad_yes', 'appet_poor','pe_yes', 'ane_yes']]
```

```python
independent
```

```python
dependent=dataset['classification_ckd\t']
```

```python
dependent
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.model_selection import GridSearchCV
param_grid={'criterion':['gini','entropy'], 'max_features':["sqrt"], 'n_estimators':[10,100]}
                    
                
grid= GridSearchCV(RandomForestClassifier(), param_grid, refit=True, verbose=3,cv=5,scoring='f1_weighted')
grid.fit(x_train,y_train)
            
```

```python
re=grid.cv_results_

grid_predictions=grid.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,grid_predictions)
from sklearn.metrics import classification_report
clf_report = classification_report(y_test,grid_predictions)
```

```python
from sklearn.metrics import f1_score
f1_macro=f1_score(y_test,grid_predictions,average='weighted')
```

```python
print("The f1_macro value for best parameter {}:".format(grid.best_params_),f1_macro)
```

```python
print("The confusion Matrix:\n",cm)
```

```python
print("clf report:\n",clf_report)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test,grid.predict_proba(x_test)[:,1])
```
